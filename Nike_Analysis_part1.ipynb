```python
# Creating a chart that compares revenue to ebit and indicating EBIT's percentage of Revenue
df['year'] = df['year'].str[:4]
df['Revenue'] = pd.to_numeric(df['Revenue'])
df['EBIT'] = pd.to_numeric(df['EBIT'])
df['EBIT_pct'] = (df['EBIT']/df["Revenue"]) * 100

plt.figure(figsize=(14,7))
bar_width = 0.4
x = range(len(df))
plt.bar(x, df['Revenue'], width=bar_width, label='Revenue', color='lightblue')
plt.bar([i+bar_width for i in x], df["EBIT"], width=bar_width, label="EBIT", color='blue')
for i, pct in enumerate(df['EBIT_pct']):
    plt.text(i+bar_width, df['EBIT'].iloc[i]+500, f'{pct:.1f}%', ha='center', va='bottom', fontsize=8)

plt.xticks([i + bar_width / 2 for i in x], df["year"], rotation=45)
plt.title("Nike Revenue vs EBIT (2009–2025) with EBIT % of Revenue")
plt.xlabel("Year")
plt.ylabel("USD (in millions)")
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
# Calculating and visualizing EBIT margin 
df["EBIT_margin"] = (df["EBIT"] / df["Revenue"]) * 100
median_margin = df["EBIT_margin"].median()

# Preparing barchart 
plt.figure(figsize=(14, 6))
bars = plt.bar(df["year"], df["EBIT_margin"], color="lightblue")

# Adding % labels above bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 0.3, f'{height:.1f}%', ha='center', va='bottom', fontsize=8)

#Median
plt.axhline(median_margin, color="blue", linestyle="--", linewidth=1.5, label=f"Median EBIT Margin: {median_margin:.1f}%")
plt.legend()

# Finalizing
plt.title("Nike EBIT Margin (%) — 2009 to 2025")
plt.xlabel("Year")
plt.ylabel("EBIT Margin (%)")
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()
```
