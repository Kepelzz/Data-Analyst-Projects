```python
df_copy=df.copy()
df_copy=df_copy.drop(df_copy.index[10472])
```

```python
df_copy[~df_copy['Reviews'].str.isnumeric()]
# Converting Review Datatype to integers 
df_copy['Reviews']=df_copy['Reviews'].astype(int)
df_copy.info()
```

```python
# Working on Size 
df_copy['Size'].unique()
# Too many values with M, so it would be better to convert to K-values
```

```python
df_copy['Size']=df_copy['Size'].str.replace('M','000')
df_copy['Size']=df_copy['Size'].str.replace('k','')
df_copy['Size']=df_copy['Size'].replace('Varies with device', np.nan)
df_copy['Size']=df_copy['Size'].astype(float)
```

```python
df_copy.info() # now it's a float value
```

```python
# Working on Installs and price columns
df_copy['Installs'].unique()
```

```python
df_copy['Price'].unique()
```

```python
# Removing special characters from both columns 
chars_to_remove=['+',',','$']
cols_to_clean=['Installs','Price']
for item in chars_to_remove:
    for cols in cols_to_clean:
        df_copy[cols]=df_copy[cols].str.replace(item,'')
```

```python
df_copy['Installs'].unique() # Installs column is clean 
```

```python
df_copy['Price'].unique() # Price column is clean as well 
```

```python
# Conversion to integer and float value 
df_copy['Installs']=df_copy['Installs'].astype('int')
df_copy['Price']=df_copy['Price'].astype('float')
df_copy.info()
```

```python
# Working on Last Updated column and converting everything to one format 
df_copy['Last Updated'].unique()
df_copy['Last Updated']=pd.to_datetime(df_copy['Last Updated'])
df_copy['Day']=df_copy['Last Updated'].dt.day
df_copy['Month']=df_copy['Last Updated'].dt.month
df_copy['Year']=df_copy['Last Updated'].dt.year
df_copy.info()
```

```python
df_copy.head()
```

```python
# EDA part 
# Observation: The dataset has duplicates that should be removed 
df_copy[df_copy.duplicated('App')].shape
```

```python
df_copy.drop_duplicates(subset=['App'], keep='first', inplace=True)
```

```python
df_copy.shape
```

```python
# Exploring data further 
# It's necessary to understand how many numerical and categorical values there are: 
numerical_features=[feature for feature in df_copy.columns if df_copy[feature].dtype != 'O']
categorical_features=[feature for feature in df_copy.columns if df_copy[feature].dtype == 'O']

#print columns
print('We have {} numerical features: {}'.format(len(numerical_features), numerical_features))
print('\nWe have {} categorical features: {}'.format(len(categorical_features), categorical_features))
```

```python
# Proportion of count data on categorical columns 
for col in categorical_features:
    print(df_copy[col].value_counts(normalize=True)*100)
    print('----------------------------')
```

```python
# Proportion of count data on numerical columns 
plt.figure(figsize=(15,15))
plt.suptitle('Univariate Analysis of Numerical Features', fontsize=20, fontweight='bold', alpha=0.8, y=1.)

for i in range(0, len(numerical_features)):
    plt.subplot(5,3, i+1)
    sns.kdeplot(x=df_copy[numerical_features[i]],shade=True, color='r')
    plt.xlabel(numerical_features[i])
    plt.tight_layout()
```

```python
# Observations: 
# Rating and year are left-side skewed while reviews, size, installs, and price are right-side skewed.
# Since data is skewed, the outliers are present 

# Categorical columns 
plt.figure(figsize=(20,15))
plt.suptitle('Univariate Analysis of Categorical Features', fontsize=20, fontweight='bold', alpha=0.8, y=1.)
category = ['Type', 'Content Rating']
for i in range(0, len(category)):
    plt.subplot(2,2, i+1)
    sns.countplot(x=df_copy[category[i]], palette='Set2')
    plt.xlabel(category[i])
    plt.xticks(rotation=45)
    plt.tight_layout()
```

```python
#Observations 
# Free has the major part of distribution 
# Everyone category has the biggest distribution 
```

```python
# Which is the most popular app category? 
df_copy['Category'].value_counts().plot.pie(y=df_copy['Category'],figsize=(15,16), autopct='%1.1f')
```

```python
#Observations: 
# Family is the most popular app category 
# Family, games and tools comparise the most categories in playstore;
# Beauty, comics, arts and weather comprise the least part of playstore.  
```

```python
# Top 10 App Categories 
category=pd.DataFrame(df_copy['Category'].value_counts()) ## DataFrame of apps on the basis of category 
category.rename(columns={'Category':'Count'},inplace=True)
```

```python
# Top 10 apps barchat 
category_df = category.reset_index()
category_df.columns = ['Category', 'Count'] # Need to convert category from a series to a dataframe 

plt.figure(figsize=(15,6))
sns.barplot(x=category.index[:10], y='Count', data=category_df[:10], palette='hls')
plt.title('Top 10 App Categories')
plt.xticks(rotation=90)
plt.show()
```

```python
# Observations: 
# Family category has the most number of apps with 18% of apps within this category. It is followed by games category which has 11% of the apps; 
# Least number of apps belong to the beauty category with less than 1% of the total apps. 
```
